```python
# 生成并保存 health / patient 的 GCN 特征

import sys
from pathlib import Path

# 自动识别项目根：在项目根或 notebook/ 下运行都能找到 src
ROOT = Path.cwd().resolve()
if not (ROOT / "src" / "msn_pipeline").exists():
    ROOT = ROOT.parent
if str(ROOT / "src") not in sys.path:
    sys.path.insert(0, str(ROOT / "src"))

print("项目根:", ROOT)
```

```python
from msn_pipeline.features.gcn import extract_all_patients
from msn_pipeline.data import DataPaths

DATA_ROOT = ROOT / "data"
RESULTS_ROOT = ROOT / "results"

paths = DataPaths(root=DATA_ROOT)
health_dir = paths.control_csv_dir()
patient_dir = paths.patient_csv_dir()

print("健康人特征目录:", health_dir)
print("患者特征目录:", patient_dir)
```

```python
# 1) 健康人 GCN 特征
health_gcn_csv = RESULTS_ROOT / "health_gcn_features.csv"
health_gcn_df = extract_all_patients(
    input_dir=health_dir,
    output_file=health_gcn_csv,
    in_channels=7,        # 和你的 feature 矩阵列数保持一致
    hidden_channels=64,
    out_channels=32,      # 最终每人 32 维 GCN 特征
    target_density=0.2,
    metric="pearson",
)
print("健康人 GCN 特征已保存到:", health_gcn_csv)
display(health_gcn_df.head())
```

```python
# 2) 患者 GCN 特征
patient_gcn_csv = RESULTS_ROOT / "patient_gcn_features.csv"
patient_gcn_df = extract_all_patients(
    input_dir=patient_dir,
    output_file=patient_gcn_csv,
    in_channels=7,
    hidden_channels=64,
    out_channels=32,
    target_density=0.2,
    metric="pearson",
)
print("患者 GCN 特征已保存到:", patient_gcn_csv)
display(patient_gcn_df.head())
```
